```python
def Grafica(nFil,nCol,data):
    fig,axes = plt.subplots(nrows = nFil, ncols = nCol)
    fig.set_size_inches(6*nFil+5,3*(nCol+2))
    sns.set_style('whitegrid')
    for item in zip(axes.ravel(),data.columns):
        axes, variable = item
        if data[variable].dtype=='object':
            sns.countplot(x=variable, data=df, ax=axes)
            axes.set_title(variable)
        else:
            sns.boxplot(y=variable, data=df, ax=axes)
            axes.set_title(variable)
    plt.show()
```

```python
Grafica(2,3,df)
```

```python
def Grafica_var_vs_target(filas,columnas,data,target):
    fig,axes = plt.subplots(nrows=filas,ncols=columnas)
    fig.set_size_inches(6*filas+5,3*(columnas+2))
    sns.set_style('whitegrid')
    for item in zip(axes.ravel(),data.columns):
        axes, variable = item
        if data[variable].dtype == 'object':
            if target !=variable:
                df_var_target=data.groupby([target,variable]).size().reset_index(name="Count")
                sns.barplot(x = target, y = "Count", hue = variable, data = df_var_target,ax =axes)
                axes.set_title(f'{target} vs Cantidad de {variable}')
        else:
            sns.swarmplot(x = target ,y = variable ,data = data ,ax = axes )
            axes.set_title(f'{target} vs {variable}')
    plt.show()
```

```python
Grafica_var_vs_target(3,2,df,"Drug")
```

# Procesamiento Limpieza y separación de Datos

```python
from sklearn.preprocessing import LabelEncoder # aqui no importa el nombre
def label_encoder(datos_categoria):
    le = LabelEncoder()
    df[datos_categoria]=le.fit_transform(df[datos_categoria])
```

```python
variables = ["Sex","BP","Cholesterol","Na_to_K","Drug"]
```

```python
for i in variables:
    label_encoder(i)
df.head()
```

```python
x = df.drop("Drug",axis=1)
y = df.Drug
x.head()
```

```python
y.head()
```

# Crear el set de entrenamiento y de prueba

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x ,y ,test_size= 0.2,random_state = 42,shuffle = True)
```

```python
y_train_y = (y_train == 0)
y_test_y = (y_test==0)
```

```python
y_train_y.head()
```

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(random_state = 42)
sgd.fit(x_train,y_train_y)
```

# evaluando el modelo

```python
sgd.predict([x_train.loc[0]]), y_train_y.loc[0]
```

```python
sgd.predict([x_train.loc[14]]), y_train_y.loc[14]
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd,x_train,y_train_y,cv=3,scoring="accuracy").mean()
```

# Matriz de confunsion

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd,x_train,y_train_y,cv=3)
y_train_pred
```

```python
#Importar confusion_matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_y,y_train_pred)
```

# Presicion y Recall

```python
#Importar precision_score y recall_score 
from sklearn.metrics import precision_score,recall_score
p = precision_score(y_train_y,y_train_pred)
r = recall_score(y_train_y,y_train_pred)

p,r
```

# Cambiamos de clasificador

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = 42)
rfc.fit(x_train,y_train_y)
y_train_pred = cross_val_predict(rfc,x_train,y_train_y,cv=3)
confusion_matrix(y_train_y,y_train_pred)
```

```python
p = precision_score(y_train_y,y_train_pred)
r = recall_score(y_train_y,y_train_pred)
p,r
```

```python
from sklearn.metrics import f1_score
f1_score(y_train_y,y_train_pred)
```

```python
from sklearn.metrics import roc_curve
y_forest = cross_val_predict(rfc,x_train,y_train_y,cv=3,method="predict_proba")
y_scores_forest = y_forest[:,1]
fpr_forest,tpr_forest,umbral_forest = roc_curve(y_train_y,y_scores_forest)
```

```python
#Graficar la curva ROC y la predicción de random forest 
plt.plot(fpr_forest, tpr_forest, label = "ROC Curve")
plt.plot(fpr_forest,tpr_forest, label = "RF ROC CURVE")
plt.plot([0,1],[0,1],"k--")
plt.xlabel("FRR - False positive Rate")
plt.ylabel("TPR - RECALL")
plt.grid()
plt.show()
```

```python
#Calcular el puntaje (área bajo la curva) de random forest  
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_y,y_scores_forest)
```
